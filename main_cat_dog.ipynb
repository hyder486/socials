```python
from skimage.io import imread
from skimage.io import imshow
from skimage.transform import resize
import os
```

# Loading and processing the images

```python
file_path = 'C:/Users/User/dataset/training_set/cats/'
images = []
labels = []
for i in os.listdir(file_path):
    image = imread(os.path.join(file_path, i))
    new_size = (64,64)
    resized_image = resize(image, new_size)
    images.append(resized_image.flatten())
    label = 0
    labels.append(label)
print(images)
print(labels)
```

```python
file_path = 'C:/Users/User/dataset/test_set/cats/'
for i in os.listdir(file_path):
    image = imread(os.path.join(file_path, i))
    new_size = (64,64)
    resized_image = resize(image, new_size)
    images.append(resized_image.flatten())
    label = 0
    labels.append(label)
print(images)
print(labels)
```

```python
cat_images = images
cat_labels = labels
```

```python
file_path = 'C:/Users/User/dataset/training_set/dogs/'
images = []
labels = []
for i in os.listdir(file_path):
    image = imread(os.path.join(file_path, i))
    new_size = (64,64)
    resized_image = resize(image, new_size)
    images.append(resized_image.flatten())
    label = 1
    labels.append(label)
print(images)
print(labels)
```

```python
file_path = 'C:/Users/User/dataset/test_set/dogs/'
for i in os.listdir(file_path):
    image = imread(os.path.join(file_path, i))
    new_size = (64,64)
    resized_image = resize(image, new_size)
    images.append(resized_image.flatten())
    label = 1
    labels.append(label)
print(images)
print(labels)
```

```python
dog_images = images
dog_labels = labels
```

```python
import numpy as np
x = np.array(cat_images + dog_images)
y = np.array(cat_labels + dog_labels)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
import seaborn as sns
import matplotlib.pyplot as plt
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
trainer =  GaussianNB()
trainer.fit(x_train, y_train)
y_pred = trainer.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
trainer = DecisionTreeClassifier(max_depth = 5, min_samples_split = 15, min_samples_leaf = 5,
                                 max_features = 'log2', max_leaf_nodes = 25, criterion = 'entropy')
trainer.fit(x_train, y_train)
y_pred = trainer.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
trainer = RandomForestClassifier(n_estimators = 950, random_state = 1, max_features = 'log2', max_depth = 9,
                                 max_leaf_nodes = 140, bootstrap = True, min_samples_split = 20,
                                 min_samples_leaf = 20, criterion = 'entropy')
trainer.fit(x_train, y_train)
y_pred = trainer.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```

# XG Boost Classifier

```python
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
pca = PCA(n_components=300)
x_train_reduced = pca.fit_transform(x_train)
x_test_reduced = pca.transform(x_test)
trainer =  XGBClassifier(n_estimators = 950, random_state = 1, learning_rate = 0.1, max_depth = 9,
                         subsample = 0.8, colsample_bytree = 0.5, min_child_weight = 3,
                         booster = 'gbtree', scale_pos_weight = 1, objective = 'binary:logistic')
trainer.fit(x_train_reduced, y_train)
y_pred = trainer.predict(x_test_reduced)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```

# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
trainer =  LogisticRegression(n_jobs = -1)
trainer.fit(x_train, y_train)
y_pred = trainer.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```

# SVM

```python
from sklearn.svm import SVC
trainer =  SVC(kernel = 'linear', C = 0.4, gamma = 0.01)
trainer.fit(x_train, y_train)
y_pred = trainer.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'coolwarm', fmt = 'd')
```
